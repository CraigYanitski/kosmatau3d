```python
def COBE_DIRBE_compare(folder, directory='/mnt/hpc_backup/yanitski/projects/pdr/KT3_history/MilkyWay/', verbose=False, old=False, save=False):
    
    obs = fits.open('/mnt/hpc_backup/yanitski/projects/pdr/observational_data/MilkyWay/COBE-DIRBE/regridded/temp/DIRBE_BAND10_ZSMA_regridded.fits')
    if verbose:
        print(folder)
        print(obs[0].shape)
    obs_data = obs[0].data
    # obs_line = np.asarray(obs[0].header['TRANSL'].split(', '))
    obs_long = np.linspace(obs[0].header['CRVAL1'] - obs[0].header['CDELT1'] * (obs[0].header['CRPIX1'] - 1),
                             obs[0].header['CRVAL1'] + obs[0].header['CDELT1'] * (obs[0].header['NAXIS1']-obs[0].header['CRPIX1']),
                             num=obs[0].header['NAXIS1'])
    # obs_vel = np.linspace(obs[0].header['CRVAL4'] - obs[0].header['CDELT4'] * (obs[0].header['CRPIX4'] - 1),
    #                         obs[0].header['CRVAL4'] + obs[0].header['CDELT4'] * (obs[0].header['NAXIS4']-obs[0].header['CRPIX4']),
    #                         num=obs[0].header['NAXIS4'])
    i_obs_lat = int(obs[0].header['CRPIX2'])
    obs_dust = obs[0].header['TRANSL'].split(', ')
    obs_sigma = fits.open('/mnt/hpc_backup/yanitski/projects/pdr/observational_data/MilkyWay/COBE-DIRBE/regridded/temp/DIRBE_BAND10_ZSMA_regridded_error.fits')[0].data
    intensity_string = r'Intensity $(K)$'
    
    model = fits.open(directory + folder + '/synthetic_intensity.fits')
    if verbose:
        print(model[2].shape)
    model_line = np.asarray(model[2].header['DUST'].split(', '))
    model_long = np.linspace(model[2].header['CRVAL2'] - model[2].header['CDELT2'] * (model[2].header['CRPIX2'] - 0.5),
                             model[2].header['CRVAL2'] + model[2].header['CDELT2'] * (model[2].header['NAXIS2']-model[2].header['CRPIX2'] - 0.5),
                             num=model[2].header['NAXIS2']) * 180 / np.pi
    # model_vel = np.linspace(model[2].header['CRVAL4'] - model[2].header['CDELT4'] * (model[2].header['CRPIX4'] - 0.5),
    #                         model[2].header['CRVAL4'] + model[2].header['CDELT4'] * (model[2].header['NAXIS4']-model[2].header['CRPIX4'] - 0.5),
    #                         num=model[2].header['NAXIS4'])
    i_model_lat = int(model[2].header['CRPIX3'])

    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    # fig.gca(xlabel=r'Galactic longitude $(^\circ)$', ylabel=r'Intensity $(\frac{nW}{m^2 \ sr})$')
    
    for i in range(len(obs_dust)):
        ax.step(obs_long, obs_data[i, i_obs_lat, :], color='xkcd:black', label='observation')
        ax.fill_between(obs_long, obs_data[i, i_obs_lat, :]-obs_sigma[i, i_obs_lat, :], obs_data[i, i_obs_lat, :]+obs_sigma[i, i_obs_lat, :], step='pre', color='xkcd:black', lw=1, alpha=0.3)
        i_dust = np.where(model_line == obs_dust[i])[0][0]
        model_data = model[2].data[i_model_lat, :, i_dust]
        ax.step(model_long, model_data, color='xkcd:maroon', alpha=1.0, label='model')
        ax.legend()
        ax.set_title(obs_dust[i] + ' Dust Intensity')
    
    ax.set_xlabel(r'Galactic longitude $(^\circ)$')
    ax.set_ylabel(intensity_string)
    
    if save:
        plt.savefig('/home/yanitski/projects/pdr/COBE-DIRBE_galactic_plane.png')
    else:
        plt.show()
    
    return
```

```python
def COBE_FIRAS_compare(folder, directory='/mnt/hpc_backup/yanitski/projects/pdr/KT3_history/MilkyWay/', verbose=False, old=False, save=False):

    obs = readsav('/mnt/hpc_backup/yanitski/projects/pdr/observational_data/MilkyWay/COBE-FIRAS/regridded/temp/craig.idl')
    linfrq = np.array([115.3, 230.5, 345.8, 424.8, 461.0, 492.2, 556.9, 576.3, 691.5, 
                       808.1, 1113, 1460, 2226, 1901, 2060, 2311, 2459, 2589, 921.8])
    i2kkms = (2.9979**3) / (linfrq**3 * 2 * 1.3806) * 10 ** 8
    obs_line = obs['line']
    obs_long = obs['long']
    obs_data = obs['amplitude']# * i2kkms
    obs_sigma = obs['sigma']# * i2kkms
    # intensity_string = r'Intensity $(\frac{nW}{m^2 \ sr})$'
    intensity_string = r'Intensity $(K \frac{km}{s})$'
    # print(obs_line)

    model = fits.open(directory + folder + '/synthetic_intensity.fits')
    model_line = np.asarray(model[1].header['SPECIES'].split(', '))
    model_long = np.linspace(model[1].header['CRVAL2'] - model[1].header['CDELT2'] * (model[1].header['CRPIX2'] - 0.5),
                             model[1].header['CRVAL2'] + model[1].header['CDELT2'] * (model[1].header['NAXIS2']-model[1].header['CRPIX2'] - 0.5),
                             num=model[1].header['NAXIS2']) * 180 / np.pi
    model_vel = np.linspace(model[1].header['CRVAL4'] - model[1].header['CDELT4'] * (model[1].header['CRPIX4'] - 0.5),
                            model[1].header['CRVAL4'] + model[1].header['CDELT4'] * (model[1].header['NAXIS4']-model[1].header['CRPIX4'] - 0.5),
                            num=model[1].header['NAXIS4'])
    i_model_lat = int(model[1].header['CRPIX3'])
    # print(model_line)
    
    if verbose:
        print(folder)

    i = np.where(obs_line == b'C^+')[0]
    fig, axes = plt.subplots(5, 4, figsize=(25, 20), sharex=True)
    # fig.gca(xlabel=r'Galactic longitude $(^\circ)$', ylabel=r'Intensity $(\frac{nW}{m^2 \ sr})$')
    axes = axes.flatten()
    axes[-1].remove()

    for i in range(obs_line.size):

        obs_line[i] = obs_line[i].decode('utf-8')
        if obs_line[i] == 'C^0 (609)':
            obs_line[i] = 'C 1'
        if obs_line[i] == 'O_0':
            obs_line[i] = 'O 2'
        elif obs_line[i] == 'C^+':
            obs_line[i] = 'C+ 1'
        elif obs_line[i] == 'CO 7-6 +C^0 (370)':
            obs_line[i] = 'CO 7'
        elif '-' in obs_line[i]:
            obs_line[i] = obs_line[i].split('-')[0]

        if i< 19:
            axes[i].step(obs_long, obs_data[:, i], color='xkcd:black', label='observation')
            axes[i].fill_between(obs_long, obs_data[:, i]-obs_sigma[:, i], obs_data[:, i]+obs_sigma[:, i], step='pre', color='xkcd:black', lw=1, alpha=0.3)
        else:
            axes[i].step(obs_long, obs_data.sum(1), color='xkcd:black')
            axes[i].fill_between(obs_long, obs_data.sum(1)-obs_sigma.sum(1), obs_data.sum(1)+obs_sigma.sum(1), step='pre', color='xkcd:black', lw=1, alpha=0.3)

        if obs_line[i] in model_line:
            if obs_line[i] == 'CO 7':
                obs_line[i] = r'CO 7$\to$6 + C 2$\to$1'
                i_spec = np.where((model_line == 'CO 7') | (model_line == 'C 2'))
                model_data_temp = model[1].data[:, i_model_lat, :, i_spec].sum(1)[0, :, :]
                # print(model_data_temp.shape)
            else:
                i_spec = np.where(model_line == obs_line[i])
                obs_line[i] = obs_line[i] + r'$\to$' + str(int(obs_line[i].split()[1])-1)
                model_data_temp = model[1].data[:, i_model_lat, :, i_spec][0, 0, :, :]# - model[2].data[:, i_model_lat, :, 0]
            if old:
                model_data = (np.trapz(model_data_temp, model_vel, axis=0) - model[2].data[0, i_model_lat, :, 0]) / i2kkms[i]
            else:
                model_data = (np.trapz(model_data_temp, model_vel, axis=0) - model[2].data[i_model_lat, :, 0]) / i2kkms[i]
            axes[i].step(model_long, model_data, color='xkcd:maroon', alpha=1.0, label='model')
            axes[i].legend()

        if i<19:
            axes[i].set_title(obs_line[i])
        else:
            axes[i].set_title('Total IR emission')

        if 'C+ 1' in obs_line[i]:
            axes[i].set_ylim([-5, 1400])
        elif 'C 1' in obs_line[i]:
            axes[i].set_ylim([-4, 14])
        elif obs_line[i] == 'O 2':
            axes[i].set_ylim([-50, 100])
        elif obs_line[i] == 'Si_0':
            axes[i].set_ylim([-50, 100])
        elif obs_line[i] == 'CH':
            axes[i].set_ylim([-50, 300])
        elif obs_line[i] == 'N^+':
            axes[i].set_ylim([-50, 300])
        else:
            axes[i].set_ylim([-5, 20])

    axes[16].set_xlabel(r'Galactic longitude $(^\circ)$')
    axes[17].set_xlabel(r'Galactic longitude $(^\circ)$')
    axes[18].set_xlabel(r'Galactic longitude $(^\circ)$')
    axes[15].set_xlabel(r'Galactic longitude $(^\circ)$')
    axes[0].set_ylabel(intensity_string)
    axes[4].set_ylabel(intensity_string)
    axes[8].set_ylabel(intensity_string)
    axes[12].set_ylabel(intensity_string)
    axes[16].set_ylabel(intensity_string)

    if save:
        plt.savefig('/home/yanitski/projects/pdr/COBE-FIRAS_galactic_plane.png')
    else:
        plt.show()
    
    return
```

```python
def survey_compare(survey='COGAL', survey_file='co1_test_regridded', lat=0, model='r400_rcmz0_uv10', header=False, verbose=False, save=False,
                   base_dir='/mnt/hpc_backup/yanitski/projects/pdr/', obs_dir='observational_data/MilkyWay/', model_dir='KT3_history/MilkyWay/'):

    obs = fits.open(base_dir + obs_dir + f'{survey}/regridded/temp/{survey_file}.fits')
    obs_lon = np.linspace(obs[0].header['CRVAL1'] - obs[0].header['CDELT1'] * (obs[0].header['CRPIX1'] - 1),
                          obs[0].header['CRVAL1'] + obs[0].header['CDELT1'] * (
                                      obs[0].header['NAXIS1'] - obs[0].header['CRPIX1']),
                          num=obs[0].header['NAXIS1'])
    obs_lat = np.linspace(obs[0].header['CRVAL2'] - obs[0].header['CDELT2'] * (obs[0].header['CRPIX2'] - 1),
                          obs[0].header['CRVAL2'] + obs[0].header['CDELT2'] * (
                                      obs[0].header['NAXIS2'] - obs[0].header['CRPIX2']),
                          num=obs[0].header['NAXIS2'])
    if survey != 'Planck':
        obs_vel = np.linspace(obs[0].header['CRVAL3'] - obs[0].header['CDELT3'] * (obs[0].header['CRPIX3'] - 1),
                              obs[0].header['CRVAL3'] + obs[0].header['CDELT3'] * (
                                          obs[0].header['NAXIS3'] - obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS3'])
    i_obs_lat = np.where(np.abs(obs_lat-lat) == np.abs(obs_lat-0).min())[0][0]
    obs_sigma = fits.open(base_dir + obs_dir + f'{survey}/regridded/temp/{survey_file}_error.fits')[0].data[i_obs_lat, :]
    obs_line = obs[0].header['TRANSL'].split(', ')
    if verbose:
        print(obs_line)
    if survey != 'Planck':
        obs_data_temp = obs[0].data[:, i_obs_lat, :]
        obs_data = np.trapz(obs_data_temp, obs_vel, axis=0)
    else:
        obs_data = obs[0].data[i_obs_lat, :]

    model_hdu = fits.open(base_dir + model_dir + f'{model}/synthetic_intensity.fits')
    model_line = np.asarray(model_hdu[1].header['SPECIES'].split(', '))
    model_dust = np.asarray(model_hdu[2].header['DUST'].split(', '))
    model_lon = np.linspace(model_hdu[1].header['CRVAL2'] - model_hdu[1].header['CDELT2'] * (model_hdu[1].header['CRPIX2'] - 0.5),
                             model_hdu[1].header['CRVAL2'] + model_hdu[1].header['CDELT2'] * (model_hdu[1].header['NAXIS2']-model_hdu[1].header['CRPIX2'] - 0.5),
                             num=model_hdu[1].header['NAXIS2']) * 180 / np.pi
    model_lat = np.linspace(model_hdu[1].header['CRVAL3'] - model_hdu[1].header['CDELT3'] * (model_hdu[1].header['CRPIX3'] - 0.5),
                             model_hdu[1].header['CRVAL3'] + model_hdu[1].header['CDELT3'] * (model_hdu[1].header['NAXIS3']-model_hdu[1].header['CRPIX3'] - 0.5),
                             num=model_hdu[1].header['NAXIS3']) * 180 / np.pi
    model_vel = np.linspace(model_hdu[1].header['CRVAL4'] - model_hdu[1].header['CDELT4'] * (model_hdu[1].header['CRPIX4'] - 0.5),
                            model_hdu[1].header['CRVAL4'] + model_hdu[1].header['CDELT4'] * (model_hdu[1].header['NAXIS4']-model_hdu[1].header['CRPIX4'] - 0.5),
                            num=model_hdu[1].header['NAXIS4'])
    i_model_lat = np.where(np.abs(model_lat-lat) == np.abs(model_lat-0).min())[0][0]
    if verbose:
        print(model_line)
        print(model_dust)
    if header:
        pprint(fits.open(base_dir + model_dir + f'{model}/species_emissivity.fits')[0].header)
    if survey != 'Planck':
        i_spec = np.where(model_line == obs_line)[0][0]
        model_data_temp = model_hdu[1].data[:, i_model_lat, :, i_spec]
        model_data = np.trapz(model_data_temp, model_vel, axis=0) - model_hdu[2].data[i_model_lat, :, :].mean(1)
        intensity_string = r'$\varpi$ (K km s$^{-1}$)'
    else:
        i_dust = np.where(model_dust == '240um')[0][0]
        model_data = model_hdu[2].data[i_model_lat, :, i_dust]
        intensity_string = r'T$_\mathrm{B}$ (K)'

    fig, ax = plt.subplots(1, 1, figsize=(10, 5))

    ax.step(obs_lon, obs_data, color='xkcd:black', label='observation')
    ax.fill_between(obs_lon, obs_data-obs_sigma, obs_data+obs_sigma, step='pre', color='xkcd:black', lw=1, alpha=0.3)
    ax.step(model_lon, model_data, color='xkcd:maroon', alpha=1.0, label='model')

    ax.legend()
    ax.set_xlabel(r'Galactic longitude $(^\circ)$')
    ax.set_ylabel(intensity_string)
    ax.set_title('{} {} integrated intensity'.format(survey, obs_line[0]))

    plt.show()
    if save:
        plt.savefig(f'/home/yanitski/projects/pdr/{survey}_galactic_plane.png')
    plt.close()
```
