```python
import numpy as np
from astropy.io import fits
from scipy.interpolate import interp2d
import matplotlib.pyplot as plt
from copy import copy
import os

def model_selection(path='/media/yanitski/4,0 TB Hard Disk/yanitski/projects/pdr/KT3_history/MilkyWay', resolution=400, 
                    f_clump=[0.5, 1.0, 2.0, 4.0], f_interclump=[1.0, 2.0], 
                    f_density=[0.5, 1.0, 2.0, 4.0], f_uv=[10, 100], 
                    transition='CO 1', PLOT=False):
    # This function will compare the Milky Way models to the position-velocity observations depending on the
    #  dataset. It will utilise the sightlines of the dataset.
    # Currently, this will look solely at the galactic plane.
    
    obs_data = []
    lon = []
    vel = []
    
    if transition == 'CO 1':
        obs = fits.open(r'/media/yanitski/4,0 TB Hard Disk/yanitski/projects/pdr/observational_data/MilkyWay/COGAL/COGAL_deep_interp_5degree.fits')
        obs_data.append(obs[0].data.sum(0))
        obs_error = 0.5
        lon.append(np.linspace(obs[0].header['CRVAL2'],
                               obs[0].header['CRVAL2']+obs[0].header['CDELT2']*(obs[0].header['NAXIS2']-1),
                               num=obs[0].header['NAXIS2']))
        vel.append(np.linspace(obs[0].header['CRVAL1'],
                               obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-1),
                               num=obs[0].header['NAXIS1']))
        
    
    elif transition == '13CO 2':
        print('This transition using SEDIGISM is still pending...')
        directory = '/media/yanitski/4,0 TB Hard Disk/yanitski/projects/pdr/observational_data/MilkyWay/SEDIGISM/'
        files = os.listdir(directory)
        for file in files:
            obs.append(fits.open(directory+file))
            obs_data.append(obs[0].data.sum(0))
            lon.append(np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-1)/2,
                                   obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-1)/2,
                                   num=obs[0].header['NAXIS1']))
            vel.append(np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-1)/2,
                                   obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-1)/2,
                                   num=obs[0].header['NAXIS3']))
        
    
    elif transition == 'CO 2':
        print('This transition using Mopra is still pending...')
        directory = '/media/yanitski/4,0 TB Hard Disk/yanitski/projects/pdr/observational_data/MilkyWay/SEDIGISM/'
        files = os.listdir(directory)
        for file in files:
            obs.append(fits.open(directory+file))
            obs_data.append(obs[0].data.sum(0))
            lon.append(np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-1)/2,
                                   obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-1)/2,
                                   num=obs[0].header['NAXIS1']))
            vel.append(np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-1)/2,
                                   obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-1)/2,
                                   num=obs[0].header['NAXIS3']))
    
    if len(obs) = 0:
        print('no data is available.')
        exit()
    
    params = []
    chi2_grid = []
    loglikelihood_grid = []
    
    # Loop through the different models
    for i_obs in range(len(obs_data)):
        for fcl in f_clump:
            for ficl in f_interclump:
                for fden in f_density:
                    for fuv in f_uv:

                        print('mass factor {}-{}, density factor {}, FUV factor {}'.format(fcl, ficl, fden, fuv/10))

                        # Check existance of model
                        dir_model = '/r{}_cm{}-{}_d{}_uv{}/channel_intensity.fits'.format(resolution, fcl, ficl, fden, fuv)
                        if not os.path.isfile(path+dir_model): continue

                        # Open the model
                        params.append([fcl, ficl, fden, fuv/10])
                        model = fits.open(path+dir_model)

                        # Locate the species transition
                        i_spec = np.where(np.asarray(model[1].header['SPECIES'].split(', ')) == transition)[0][0]

                        # Locate the galactic disk
                        if model[1].header['CRPIX3'] == 0:
                            i_lat = np.floor(model[1].header['NAXIS3']/2).astype(np.int)
                        else:
                            lat = np.linspace(model[1].header['CRVAL3']-model[1].header['CDELT3']*model[1].header['CRPIX3'],
                                              model[1].header['CRVAL3']+model[1].header['CDELT3']*model[1].header['CRPIX3'],
                                              num=model[1].header['NAXIS3'])
                            i_lat = np.where(np.abs(lat-0) == np.abs(lat-0).min())[0][0]

                        # Create arrays for the longitude and velocity axes
                        lon_model = np.linspace(model[1].header['CRVAL2']-model[1].header['CDELT2']*model[1].header['CRPIX2'],
                                                model[1].header['CRVAL2']+model[1].header['CDELT2']*model[1].header['CRPIX2'],
                                                num=model[1].header['NAXIS2']) * 180/np.pi
                        vel_model = np.linspace(model[1].header['CRVAL4']-model[1].header['CDELT4']*model[1].header['CRPIX4'],
                                                model[1].header['CRVAL4']+model[1].header['CDELT4']*model[1].header['CRPIX4'],
                                                num=model[1].header['NAXIS4'])

                        # Interpolate at the observational axes
                        # print('grid shape:', model[1].data[:, i_lat, :, i_spec].shape)
                        # print('axes shape:', vel_model.size, lon_model.size)
                        f_model_interp = interp2d(vel_model, lon_model, model[1].data[:,i_lat,:,i_spec].T, kind='cubic')
                        model_interp = f_model_interp(vel[i_obs], lon[i_obs])

                        # Calculate the chi**2 and overall likelihood
                        chi2 = (np.log10(obs_data[i_obs])-np.log10(model_interp))**2/obs_error**2
                        likelihood = np.exp(-chi2/2)/np.sqrt(2*np.pi*obs_error)

                        #print(model_interp)
                        #print(obs_data[~np.isnan(obs_data)])
                        #print(chi2)
                        chi2 = np.nan_to_num(chi2, nan=0)
                        likelihood = np.nan_to_num(likelihood, nan=1)
                        likelihood[likelihood == 0] = 1e-100
                        #input()

                        chi2_grid.append(chi2.sum())
                        loglikelihood_grid.append(np.log10(likelihood).sum())
                        print('  ', likelihood.min())

                        if PLOT:
                            lon_grid,vel_grid = np.meshgrid(lon[i_obs], vel[i_obs])
                            i_nan = np.isnan(obs_data[i_obs])
                            fig,ax = plt.subplots(1, 1, figsize=(7,7))
                            #cm = ax.pcolormesh(lon_grid.T, vel_grid.T, np.log10(model_interp))
                            #cb = fig.colorbar(cm, ax=ax, fraction=0.02)
                            ax.contour(lon_grid.T, vel_grid.T, np.log10(obs_data[i_obs]))
                            ax.set_xlabel(r'Longitude ($^\circ$)', fontsize=16)
                            ax.set_ylabel(r'Radial Velocity ($\frac{km}{s}$)', fontsize=16)
                            ax.invert_xaxis()
                            fig.savefig('synthetic_observation.png')
                    
    print('\n\n')
    
    i_bestfit1 = np.where(chi2_grid == np.min(chi2_grid))[0][0]
    i_bestfit2 = np.where(loglikelihood_grid == np.max(loglikelihood_grid))[0][0]
    
    print('The best-fitting model has parameters f_cl {}, f_icl {}, f_dens {}, and f_FUV {}'.format(params[i_bestfit1][0], params[i_bestfit1][1], params[i_bestfit1][2], params[i_bestfit1][3]))
    print('The best-fitting model has parameters f_cl {}, f_icl {}, f_dens {}, and f_FUV {}'.format(params[i_bestfit2][0], params[i_bestfit2][1], params[i_bestfit2][2], params[i_bestfit2][3]))
    
    return chi2_grid,loglikelihood_grid,params
```

```python
chi2,likelihood,params = model_selection(f_clump=[1.0], f_interclump=[1.0], f_density=[1.0], f_uv=[10], PLOT=True)
```
