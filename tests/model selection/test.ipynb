```python
import numpy as np
from astropy.io import fits
from scipy.interpolate import interp2d
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from functools import lru_cache
from copy import copy
import os


def model_selection(path='/media/4,0 TB Hard Disk/yanitski/projects/pdr/KT3_history/MilkyWay', resolution=400, 
                    f_clump=[0.5, 1.0, 2.0, 4.0], f_interclump=[1.0, 2.0], 
                    f_density=[0.5, 1.0, 2.0, 4.0], f_uv=[10, 100], 
                    transition='CO 1', mission='', PLOT=False, PRINT=False):
    # This function will compare the Milky Way models to the position-velocity observations depending on the
    #  dataset. It will utilise the sightlines of the dataset.
    # Currently, this will look solely at the galactic plane.
    
    obs_data = []
    lon = []
    vel = []
    
    if path[-1] != '/': path += '/'
    directory = path.replace('KT3_history', 'observational_data')+mission
    if directory[-1] != '/': directory += '/'
    files = os.listdir(directory)
    
    if len(files) == 0:
        print('no data is available.')
        exit()
    
    # Loop through the different models
    # for i_obs in range(len(obs_data)):
    for file in files:
        if mission == 'COGAL' and 'interp' in file:
            transitions = ['CO 1']
            obs = fits.open(directory+file)
            obs_data = np.nansum(obs[0].data, axis=0)
            obs_error = 0.5
            lon = np.linspace(obs[0].header['CRVAL2']-obs[0].header['CDELT2']*(obs[0].header['CRPIX2']-1),
                              obs[0].header['CRVAL2']+obs[0].header['CDELT2']*(obs[0].header['NAXIS2']-obs[0].header['CRPIX2']),
                              num=obs[0].header['NAXIS2'])
            lat = np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['CRPIX3']-1),
                              obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS3'])
            vel = np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['CRPIX1']-1),
                              obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS1'])
            vmin = obs[0].header['DATAMIN']
            vmax = obs[0].header['DATAMAX']
        elif mission == 'Mopra' and ('_Vfull' in file):
            if '12CO' in file.split('_')[0]:
                transitions = ['CO 1']
            elif '13CO' in file.split('_')[0]:
                transitions = ['13CO 1']
            obs = fits.open(directory+file)
            err = fits.open(directory+file.replace('_Vfull', '.sigma'))
            obs_data = np.nansum(obs[0].data, axis=1)
            obs_error = np.nansum(err[0].data, axis=1)
            lon = np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['CRPIX1']-1),
                              obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-obs[0].header['CRPIX1']),
                              num=obs[0].header['NAXIS1'])
            lat = np.linspace(obs[0].header['CRVAL2']-obs[0].header['CDELT2']*(obs[0].header['CRPIX2']-1),
                              obs[0].header['CRVAL2']+obs[0].header['CDELT2']*(obs[0].header['NAXIS2']-obs[0].header['CRPIX2']),
                              num=obs[0].header['NAXIS2'])
            vel = np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['CRPIX3']-1),
                              obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS3'])
            vmin = obs[0].header['DATAMIN']
            vmax = obs[0].header['DATAMAX']
        elif (mission == 'ThrUMMS'):
            if '12' in file.split('.')[-2]:
                transitions = ['CO 1']
            elif '13' in file.split('.')[-2]:
                transitions = ['13CO 1']
            obs = fits.open(directory+file)
            #err = fits.open(directory+file.replace('_Vfull', '.sigma'))
            obs_data = np.nansum(obs[0].data, axis=1)
            obs_error = 1.5 #err[0].data.sum(0)
            lon = np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['CRPIX1']-1),
                              obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-obs[0].header['CRPIX1']),
                              num=obs[0].header['NAXIS1'])
            lat = np.linspace(obs[0].header['CRVAL2']-obs[0].header['CDELT2']*(obs[0].header['CRPIX2']-1),
                              obs[0].header['CRVAL2']+obs[0].header['CDELT2']*(obs[0].header['NAXIS2']-obs[0].header['CRPIX2']),
                              num=obs[0].header['NAXIS2'])
            vel = np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['CRPIX3']-1),
                              obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS3'])
            vmin = obs[0].header['DATAMIN']
            vmax = obs[0].header['DATAMAX']
        elif mission == 'SEDIGISM':
            transitions = ['13CO 2']
            obs = fits.open(directory+file)
            obs_data = np.nansum(obs[0].data, axis=1)
            obs_error = 0.95 # obs[0].header['SIGMA']
            lon = np.linspace(obs[0].header['CRVAL1']-obs[0].header['CDELT1']*(obs[0].header['CRPIX1']-1),
                              obs[0].header['CRVAL1']+obs[0].header['CDELT1']*(obs[0].header['NAXIS1']-obs[0].header['CRPIX1']),
                              num=obs[0].header['NAXIS1'])
            lat = np.linspace(obs[0].header['CRVAL2']-obs[0].header['CDELT2']*(obs[0].header['CRPIX2']-1),
                              obs[0].header['CRVAL2']+obs[0].header['CDELT2']*(obs[0].header['NAXIS2']-obs[0].header['CRPIX2']),
                              num=obs[0].header['NAXIS2'])
            vel = np.linspace(obs[0].header['CRVAL3']-obs[0].header['CDELT3']*(obs[0].header['CRPIX3']-1),
                              obs[0].header['CRVAL3']+obs[0].header['CDELT3']*(obs[0].header['NAXIS3']-obs[0].header['CRPIX3']),
                              num=obs[0].header['NAXIS3'])
            vmin = obs[0].header['DATAMIN']
            vmax = obs[0].header['DATAMAX']
        elif mission == 'COBE' or mission == 'COBE-FIRAS':
            if 'HIGH' in file.split('.')[0]:
                transitions = ['CO 6', 'C 2', 'C+ 1', 'O 2']
                transition_indeces = [0, 1, 5, 6]
            elif 'HRES' in file.split('.')[0]:
                transitions = ['CO 1', 'CO 2', 'CO 3', 'CO 4', 'C 1', 'CO 5']
                transition_indeces = [0, 1, 2, 4, 5, 7]
            elif 'LOWF' in file.split('.')[0]:
                transitions = ['CO 1', 'CO 2', 'CO 3', 'CO 4', 'C 1', 'CO 5']
                transition_indeces = [0, 1, 2, 4, 5, 7]
            obs = fits.open(directory+file)
            linfrq = np.array([obs[0].header[key] for key in obs[0].header.keys() if 'LINFRQ' in key]) * 1e9
            obs_data = obs[0].data['LINE_FLU']*2.9979**3/linfreq**3/2/1.38*10**11
            obs_error = obs[0].data['LINE_FL2']*2.9979**3/linfreq**3/2/1.38*10**11
            lon = obs[1].data['GAL_LON']
            lat = obs[1].data['GAL_LAT']
        else:
            continue
        
        chi2,loglikelihood,params = compare(mission=mission, transitions=transitions, PRINT=PRINT, path=path, 
                                            plot=plot, resolution=resolution, f_clump=f_clump, 
                                            f_interclump=f_interclump, f_density=f_density, f_uv=f_uv)
        
    return

def compare(**kwargs):
        
    for i, transition in enumerate(transitions):

        chi2_grid = []
        loglikelihood_grid = []
        params = []

        if os.path.isdir(path+'fit_results/{}/{}/{}/'.format(mission, file.replace('.fits', ''), transition)) == False:
            os.makedirs(path+'fit_results/{}/{}/{}/'.format(mission, file.replace('.fits', ''), transition))
        for fcl in f_clump:
            for ficl in f_interclump:
                for fden in f_density:
                    for fuv in f_uv:

                        if PRINT:
                            print('mass factor {}-{}, density factor {}, FUV factor {}'.format(fcl, ficl, fden, fuv/10))

                        # Check existance of model
                        dir_model = '/r{}_cm{}-{}_d{}_uv{}/channel_intensity.fits'.format(resolution, fcl, ficl, fden, fuv)
                        if not os.path.isfile(path+dir_model): continue

                        # Open the model
                        params.append([fcl, ficl, fden, fuv/10])
                        model = fits.open(path+dir_model)

                        # Locate the species transition
                        i_spec = np.where(np.asarray(model[1].header['SPECIES'].split(', ')) == transition)[0][0]

#                         # Locate the galactic disk
#                         if model[1].header['CRPIX3'] == 0:
#                             i_lat = np.floor(model[1].header['NAXIS3']/2).astype(np.int)
#                         else:

                        # Create arrays for the longitude and velocity axes
                        lat_model = np.linspace(model[1].header['CRVAL3']-model[1].header['CDELT3']*model[1].header['CRPIX3'],
                                                model[1].header['CRVAL3']+model[1].header['CDELT3']*(model[1].header['NAXIS3']-model[1].header['CRPIX3']),
                                                num=model[1].header['NAXIS3'])
                        lon_model = np.linspace(model[1].header['CRVAL2']-model[1].header['CDELT2']*model[1].header['CRPIX2'],
                                                model[1].header['CRVAL2']+model[1].header['CDELT2']*model[1].header['CRPIX2'],
                                                num=model[1].header['NAXIS2']) * 180/np.pi
                        lon_model[lon_model<0] += 360
                        vel_model = np.linspace(model[1].header['CRVAL4']-model[1].header['CDELT4']*model[1].header['CRPIX4'],
                                                model[1].header['CRVAL4']+model[1].header['CDELT4']*model[1].header['CRPIX4'],
                                                num=model[1].header['NAXIS4'])

                        # Identify the midplane
                        i_lat = np.where(np.abs(lat_model-0) == np.abs(lat_model-0).min())[0][0]

                        # Interpolate at the observational axes
                        # print('grid shape:', model[1].data[:, i_lat, :, i_spec].shape)
                        # print('axes shape:', vel_model.size, lon_model.size)
                        if mission == 'COBE-FIRAS':
                            model_int = trap(model[1].data, vel_model, axis=0)
                            f_model_interp = interp2d(lon_model, lat_model, model_int[:, :, i_spec])
                            model_interp = f_model_interp(lon, lat)
                        else:
                            f_model_interp = interp2d(vel_model, lon_model, model[1].data[:,i_lat,:,i_spec].T, kind='cubic')

                            model_interp = f_model_interp(vel, lon)

                        # Calculate the chi**2 and overall likelihood
                        chi2 = (obs_data-model_interp.T)**2/obs_error**2
                        likelihood = np.exp(-chi2/2)/np.sqrt(2*np.pi*obs_error)

                        #print(model_interp)
                        #print(obs_data[~np.isnan(obs_data)])
                        #print(chi2)
                        chi2 = np.nan_to_num(chi2, nan=0)
                        likelihood = np.nan_to_num(likelihood, nan=1)
                        likelihood[likelihood == 0] = 1e-100
                        #input()

                        chi2_grid.append(chi2.sum())
                        loglikelihood_grid.append(np.log10(likelihood).sum())
#                             print('  ', likelihood.min())

                        # print(obs_data.min(), obs_data.max())

                        if PLOT:
                            if mission == 'COBE-FIRAS':
                                continue
                            else:
                                lon_grid,vel_grid = np.meshgrid(lon, vel)
                                i_nan = np.isnan(obs_data)
                                fig,ax = plt.subplots(1, 1, figsize=(7,7))
                                #cm = ax.pcolormesh(lon_grid.T, vel_grid.T, np.log10(model_interp))
                                cm = ax.pcolormesh(lon_grid, vel_grid, obs_data, norm=colors.SymLogNorm(base=10, linthresh=0.1, vmin=vmin, vmax=vmax), shading='gouraud')
                                cb = fig.colorbar(cm, ax=ax, fraction=0.02)
                                ax.set_xlabel(r'Longitude ($^\circ$)', fontsize=16)
                                ax.set_ylabel(r'Radial Velocity ($\frac{km}{s}$)', fontsize=16)
                                ax.invert_xaxis()
                            fig.savefig(path+'fit_results/{}/{}/{}/synthetic_observation_cm{}-{}_d{}_fuv{}.png'.format(mission, file.replace('.fits', ''), transition, fcl, ficl, fden, fuv))
                            plt.close()

                        model.close()
                        del f_model_interp
                        del model_interp
                        del chi2
                        del likelihood
                        del lon_model
                        del lat_model
                        del vel_model

        i_bestfit2 = np.where(loglikelihood_grid == np.max(loglikelihood_grid))[0][0]
        # print(i_bestfit2)
        print('The best-fitting model for {}\n  has parameters f_cl {}, f_icl {}, f_dens {}, and f_FUV {}'.format(file.replace('.fits', ''), params[i_bestfit2][0], params[i_bestfit2][1], params[i_bestfit2][2], params[i_bestfit2][3]))

        np.save(path+'fit_results/{}/{}/{}/chi2.npy'.format(mission, file.replace('.fits', ''), transition), chi2_grid)
        np.save(path+'fit_results/{}/{}/{}/loglikelihood.npy'.format(mission, file.replace('.fits', ''), transition), loglikelihood_grid)
        np.save(path+'fit_results/{}/{}/{}/parameters.npy'.format(mission, file.replace('.fits', ''), transition), params)

    obs.close()
    del obs_data
    del obs_error
    del lon
    del lat
    del vel
    
    print('\n\n')
    
    i_bestfit1 = np.where(chi2_grid == np.min(chi2_grid))[0][0]
    i_bestfit2 = np.where(loglikelihood_grid == np.max(loglikelihood_grid))[0][0]
    
    print('The best-fitting model has parameters f_cl {}, f_icl {}, f_dens {}, and f_FUV {}'.format(params[i_bestfit1][0], params[i_bestfit1][1], params[i_bestfit1][2], params[i_bestfit1][3]))
    print('The best-fitting model has parameters f_cl {}, f_icl {}, f_dens {}, and f_FUV {}'.format(params[i_bestfit2][0], params[i_bestfit2][1], params[i_bestfit2][2], params[i_bestfit2][3]))
    
    return
```
